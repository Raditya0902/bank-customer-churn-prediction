```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.drop(["RowNumber", "CustomerId", "Surname"], axis=1,inplace=True)
df.head()
```

```python
df.Gender.replace({"Female":1, "Male": 0}, inplace=True)
df.head()
```

```python
df.Geography.unique()
```

```python
geo = pd.get_dummies(df['Geography'],drop_first=True).astype(int)
df.drop(["Geography"], axis=1, inplace=True)
df = pd.concat([df,geo], axis=1)
df.head()
```

```python
df.NumOfProducts.unique()
```

```python
cols_to_scale = ['CreditScore','Age','Tenure','Balance', 'NumOfProducts', 'EstimatedSalary']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
df.dtypes
```

```python
y = df.Exited
```

```python
y
```

```python
X = df.drop("Exited", axis=1)
```

```python
X
```

```python
print(len(X))
print(len(X.columns))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([keras.layers.Dense(9,input_shape=(11,), activation='relu'),
                          keras.layers.Dense(6,activation='relu'),
                         keras.layers.Dense(1, activation='sigmoid')])
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
yp = []
for ele in y_pred:
    if ele > 0.5:
        yp.append(1)
    else:
        yp.append(0)
```

```python
yp[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, yp))
```

```python
confusion_matrix(y_test,yp)
```

```python
#Below code is for undersampling, a technique used for skewed datasets.
```

```python
class_0, class_1 = df.Exited.value_counts()
```

```python
class_0, class_1
```

```python
df_class_0 = df[df.Exited == 0]
df_class_1 = df[df.Exited == 1]
```

```python
df_under_class_0 = df_class_0.sample(class_1)
df_under_class_0.head()
```

```python
print(df_under_class_0.shape,df_class_1.shape)
df_class_1.head()
```

```python
under_sampling_df = pd.concat([df_under_class_0,df_class_1], axis=0)
```

```python
under_sampling_df.head()
under_sampling_df.shape
```

```python
y = under_sampling_df.Exited
X = under_sampling_df.drop("Exited", axis=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([keras.layers.Dense(9,input_shape=(11,), activation='relu'),
                          keras.layers.Dense(6,activation='relu'),
                         keras.layers.Dense(1, activation='sigmoid')])
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
yp = []
for ele in y_pred:
    if ele > 0.5:
        yp.append(1)
    else:
        yp.append(0)
```

```python
yp[:10]
```

```python
print(classification_report(y_test, yp))
```

```python
confusion_matrix(y_test,yp)
```

```python
#SMOTE code
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='minority')
```

```python
X = df.drop(['Exited'],axis='columns')
y = df['Exited']
```

```python
X_smote, y_smote = smote.fit_resample(X,y)
```

```python
y_smote.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_smote,y_smote,test_size=0.2,stratify=y_smote)
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([keras.layers.Dense(9,input_shape=(11,), activation='relu'),
                          keras.layers.Dense(6,activation='relu'),
                         keras.layers.Dense(1, activation='sigmoid')])
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
yp = []
for ele in y_pred:
    if ele > 0.5:
        yp.append(1)
    else:
        yp.append(0)
```

```python
yp[:10]
```

```python
print(classification_report(y_test, yp))
```

```python
confusion_matrix(y_test,yp)
```
